```python
import os
from pathlib import Path
import pandas as pd

pd.set_option('display.max_columns', None)

# Definir caminho para o arquivo parquet local
# O notebook está em staging/parquet, então subimos 1 nível e entramos em data
current_dir = Path.cwd()
staging_data_path = current_dir.parent / "data" / "mapeamentos_titulados_2023.parquet"

print(f"Diretório atual: {current_dir}")
print(f"Lendo arquivo: {staging_data_path}")
print(f"Arquivo existe: {staging_data_path.exists()}")

# Ler o arquivo parquet local
if staging_data_path.exists():
    df = pd.read_parquet(staging_data_path)
    print(f"✅ Arquivo carregado com sucesso! {len(df):,} registros")
else:
    print(f"❌ Arquivo não encontrado!")
```
